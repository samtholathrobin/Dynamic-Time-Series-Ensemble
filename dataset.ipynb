```python
import yfinance as yahooFinance
import pandas as pd
GetInformation = yahooFinance.Ticker("GOOG")

# max->maximum number of daily prices available for the stock.
# Valid options are 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y and ytd.

df=GetInformation.history(period="max")
df.head()
```

```python
df = df.reset_index()
df.drop(columns=['Open', 'High', 'Low', 'Volume','Dividends','Stock Splits'],inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df['Date']=df['Date'].dt.date
df.shape
```

```python
df.to_csv('Google stocks.csv',index=False)
```
